```python
# Retrieve page with the requests module
html = requests.get(url1)

# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info and input info to list
results = soup.find_all('span', class_='Winner')
winners = []
for result in results:
    winners.append(result.text)
```

```python
df1 = pd.DataFrame(winners, columns = ['2007 WINNERS'])
df1 = df1.reset_index(drop=True)
df1 = df1.drop([0,1,2,3])
df1 = df1.drop(df1[df1.index>27].index)
df1 = df1.reset_index(drop=True)
```
